##🔹 8. Data Visualization (Graphs)

```python
# Convert date column
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])

plt.figure()
plt.plot(df['Sale_Date'], df['Quantity_Sold'])
plt.title("Sales Over Time")
plt.xlabel("Sale_Date")
plt.ylabel("Quantity_Sold")
plt.show()
```

```python
plt.figure()
plt.hist(df['Sales_Amount'])
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Quantity_Sold'])
plt.title("Boxplot of Sales")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.set_index('Sale_Date', inplace=True)

monthly_sales = df['Quantity_Sold'].resample('M').mean()

monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.show()
```
